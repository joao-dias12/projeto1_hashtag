```python
#importar bibliotecas
import pandas as pd
```

```python
#importar base de dados
emails = pd.read_excel(r'C:\Users\João\Documents\Projeto AutomacaoIndicadores\Projeto AutomacaoIndicadores\Bases de Dados\Emails.xlsx')
lojas = pd.read_csv(r'C:\Users\João\Documents\Projeto AutomacaoIndicadores\Projeto AutomacaoIndicadores\Bases de Dados\lojas.csv', encoding='latin1', sep=';')
vendas = pd.read_excel(r'C:\Users\João\Documents\Projeto AutomacaoIndicadores\Projeto AutomacaoIndicadores\Bases de Dados\vendas.xlsx')

display(emails)
display(lojas)
display(vendas)
```

Criar uma tabela para cada loja

```python
#incluir nome da loja em vendas
vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)
```

```python
#criando uma tabela para cada loja
tabela_lojas = {} #dicionario para armazenar as lojas
for loja in lojas['Loja']: #for para percorrer cada linha da tabela de lojas
    tabela_lojas[loja]  = vendas.loc[vendas['Loja'] == loja, : ] #criando uma tabela para cada loja

display(tabela_lojas['Iguatemi Esplanada'])
display(tabela_lojas['Norte Shopping'])
display(tabela_lojas['Shopping União de Osasco'])
    
```
